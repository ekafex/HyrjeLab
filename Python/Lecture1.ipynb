# Scientific‑Python Sanity Check

Welcome! This notebook verifies that your environment is ready and gives a lightning‑fast refresher on Python syntax compared with C/C++. Work through each cell before the first live session.


## 1 · Python ↔ C Cheat Sheet

| Concept | In C / C++ | In Python |
|---------|-----------|-----------|
| Import / includes | `#include <stdio.h>` | `import numpy as np` |
| Entry point | `int main(void){}` | script executes top‑level, or `if __name__ == "__main__":` |
| Array (static) | `double a[10];` | `import numpy as np; a = np.zeros(10)` |
| Loop | `for(int i=0;i<n;i++){}` | `for i in range(n):` |
| Function | `double f(double x){return x*x;}` | `def f(x): return x**2` |
| Print | `printf("x=%f\n", x);` | `print(f"x={x}")` |
| Compile & run | `gcc prog.c` → `./a.out` | interpreted / REPL / Jupyter |


````python
import sys, platform
import numpy as np
import matplotlib
import pandas as pd
import scipy

print(f"Python  : {sys.version.split()[0]}")
print(f"Platform: {platform.platform()}")
print(f"numpy   : {np.__version__}")
print(f"matplotlib: {matplotlib.__version__}")
print(f"pandas  : {pd.__version__}")
print(f"scipy   : {scipy.__version__}")
````

````python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 500)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')
ax.set_title('Sanity‑check plot')
plt.show()
````
